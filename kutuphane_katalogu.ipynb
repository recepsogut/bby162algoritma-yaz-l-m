```
from google.colab import drive
drive.mount("/gdrive",force_remount=True)

dosya = "/gdrive/My Drive/Colab Notebooks/bby162/veri_listesi.txt"


class Catalog:
    def __init__(self):
        kıtaplar = []
        verı = open(dosya,"r",encoding = "cp1254")
        for line in verı.read().strip().split("\n"):
            kıtap = line.strip().split("|")
            kıtaplar += [kıtap]
        verı.close()
        self.kıtaplar = kıtaplar
        kıtaplar = None

    def search(self):
        anahtarkelıme = input("Aramak istediğiniz anahtar kelimeyi giriniz:")#
        output = ""

        for kıtap in self.kıtaplar:
            if len(self.kıtaplar) > 0 and anahtarkelıme in kıtap:
                output += f'Eser adı: {kıtap[1]}\n'
                output += f'Yazar: {kıtap[0]}\n'
                output += f'Yayın yılı: {kıtap[2]}\n'
                output += f'Yayıncı: {kıtap[3]}\n'
                print(output)
        if output == "":
            print("Aradığınız eser bulunamadı.")#

    def update(self):
        ısım = input("Eser adını giriniz: ")
        yazar = input("Yazarını giriniz: ")
        yıl = input("Yayın yılını giriniz: ")
        yayıncı = input("Yayınevini giriniz: ")
        newBook = f'{yazar}|{ısım}|{yıl}|{yayınevı}'
        verı = open(dosya,"a",encoding = "cp1254")
        verı.write(f'\n{newBook}')
        verı.close()
        self.kıtaplar.append([ısım,yazar,yıl,yayınevı])

    def delete(self):
        anahtarkelıme = input("Silmek istediğiniz eserin bir bilgisini giriniz: ")
        if len(self.kıtaplar) < 1:
            return None

        for kıtap in self.kıtaplar:
            if anahtarkelıme in kıtap:
                self.kıtaplar.remove(kıtap)
                break

        verı = open(dosya,"a",encoding = "cp1254")
        verı.truncate(0)
        for kıtap in self.kıtaplar:
            verı.write(f'{kıtap[1]}|{kıtap[0]}|{kıtap[2]}|{kıtap[3]}')
            if not kıtap == self.kıtaplar[-1]:
                verı.write("\n")
        verı.close()

    def listing(self):
        if len(self.kıtaplar) < 1:
            print("Eser kaydı bulunmamaktadır. ")
            return None

        output = ""
        for kıtap in self.kıtaplar:
            output += f'Eser adı: {kıtap[1]}\n'
            output += f'Yazar: {kıtap[0]}\n'
            output += f'Yayın yılı: {kıtap[2]}\n'
            output += f'Yayıncı: {kıtap[3]}\n'
            output += ("-"*60 + "\n")
        print(output)

    def main(self):
        while True:
            operasyon = input("1-Arama\n2-Eser Ekleme\n3-Eser Silme\n4-Eserleri Listeleme\n0-Çık\nYapmak istediğiniz işlemi seçiniz: ")
            if operasyon == "1":
                self.search()
            elif operasyon == "2":
                self.update()
            elif operasyon == "3":
                self.delete()
            elif operasyon == "4":
                self.listing()
            elif operasyon == "0":
                break
            else:
                print("Geçersiz istek.")

Catalog().main()
```
